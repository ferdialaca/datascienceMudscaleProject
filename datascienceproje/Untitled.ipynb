```python
plt.figure(figsize = (10,10))
sns.barplot(merged_df["Publisher"].value_counts().index[:10], merged_df["Publisher"].value_counts().iloc[:10]
           ,palette='mako'
           ,edgecolor='black'
           )
plt.xticks(rotation=90)
plt.title("Most Games produced in Specific Gaming Publisher",fontsize=20)
plt.show()
```

```python
comp_genre = merged_df[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_map = comp_genre.groupby(by=['Genre']).sum()
comp_table = comp_map.reset_index()
comp_table = pd.melt(comp_table, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
plt.figure(figsize=(15, 10))
sns.barplot(x='Genre', y='Sale_Price', hue='Sale_Area', data=comp_table)
```

```python
#We see that video sales in North America are higher for all genres 
#than in other regions. In addition, action and sports games were sold
#more than other genres.
```

```python
top_sale_reg = merged_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
top_sale_reg = top_sale_reg.sum().reset_index()
top_sale_reg = top_sale_reg.rename(columns={"index": "region", 0: "sale"})
top_sale_reg
```

```python
labels = top_sale_reg['region']
sizes = top_sale_reg['sale']
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
```

```python
#North America has the largest share of sales
```
